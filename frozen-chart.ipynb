```python
import matplotlib.pyplot as plt

# Data
shots = ['1-shot', '5-shot', '10-shot']
frozen_head = [48.7, 62.1, 69.8]
full_finetune = [51.2, 64.8, 72.5]

# Bar width and position
bar_width = 0.35
x = range(len(shots))

# Plotting
plt.figure(figsize=(8, 5))
plt.bar([p - bar_width/2 for p in x], frozen_head, width=bar_width, label='Frozen Head', color='#1f77b4')
plt.bar([p + bar_width/2 for p in x], full_finetune, width=bar_width, label='Full Finetune', color='#ff7f0e')
plt.xticks(ticks=x, labels=shots)
plt.ylabel('Accuracy (%)')
plt.title('Few-shot Accuracy Comparison')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
